```python
#import data
df_SF_Crime = pd.read_csv("https://cocl.us/sanfran_crime_dataset")
df_SF_Crime.head()
```

```python
#Data Wrangling
df_SF_Crime.dropna(subset=['PdDistrict'],axis=0,inplace=True)
df_Sf_CrimeReg = df_SF_Crime.groupby('PdDistrict',axis=0)[['IncidntNum']].count()
df_Sf_CrimeReg.sort_values(by='IncidntNum',ascending = False,inplace=True)
df_Sf_CrimeReg.reset_index(level='PdDistrict', inplace=True)
df_Sf_CrimeReg.rename(columns={'PdDistrict':'Neighborhood','IncidntNum':'Count'}, inplace=True)
df_Sf_CrimeReg
```

```python
#!wget --quiet https://cocl.us/sanfran_geojson -O san-francisco.json
# !wget --quiet https://cocl.us/sanfran_geojson/san-francisco.geojson -o san-francisco.geo.json
# world_geo = r'https://cocl.us/sanfran_geojson/san-francisco.geo.json'
world_geo =r'san-francisco.geo.json'
```

```python
# SF_geo = r'https://cocl.us/sanfran_geojson/san-francisco.geo.json' # san francisco json file
# create a SF map
SF_map = folium.Map(location = [37.7749,-122.419], zoom_start=2)
```

```python
# generate choropleth map 
SF_map.choropleth(
    geo_data=world_geo,
    data=df_Sf_CrimeReg,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='SF Crime data across Neighborhood'
)

# display map
SF_map
```
